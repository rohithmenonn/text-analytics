```python
import urllib.request
fp = urllib.request.urlopen("https://en.wikipedia.org/wiki/Internet")
mybytes = fp.read()
html_source = mybytes.decode("utf8")
fp.close()
print(html_source)
```

```python
import re
data = re.findall('src="([^"]+[\.png$])"', html_source)
print(data)
```

```python
def download_images(root_url, image_url_list):
    file_count = 1
    for url in image_url_list:
        url = root_url + url
        urllib.request.urlretrieve(url, "images/"+str(file_count)+".png")
        file_count = file_count + 1
    print("ALL images have been downloaded!")
```

```python
download_images(root_url, image_url_list)
```
